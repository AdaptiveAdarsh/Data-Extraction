```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the website to scrape
url =  'https://www.livemint.com/market/live-blog/jio-financial-services-share-price-live-blog-for-04-mar-2024-11709519715779.html'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract data using BeautifulSoup
    # (Replace this with the actual code to extract your data)
    data = []

    # Example: Extracting data from a hypothetical table
    table = soup.find('table')
    if table:
        rows = table.find_all('tr')
        header = [col.text.strip() for col in rows[0].find_all(['th'])]
        for row in rows[1:]:
            cols = row.find_all(['td', 'th'])
            row_data = [col.text.strip() for col in cols]
            data.append(row_data)

        # Create a DataFrame from the extracted data
        df = pd.DataFrame(data, columns=header)
        

        # Print the DataFrame
        print(df)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
# Convert 'Column1' to float
df['Latest Price'] = df['Latest Price'].astype(float)
df['Change'] = df['Change'].astype(float)
df['Change'] = df['Change'].astype(float)
df['52W High'] = df['52W High'].astype(float)
df['52W Low'] = df['52W Low'].astype(float)
df['Mkt. Cap'] = df['Mkt. Cap'].astype(float)
```
